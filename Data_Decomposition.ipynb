```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

## JP Morgan Stock Price

```python
# Detecting Trend Using a Hodrick-Prescott Filter
from statsmodels.tsa.filters.hp_filter import hpfilter

# df = pd.read_csv(r'India_Exchange_Rate_Dataset.xls', index_col=0,parse_dates=True)
df = pd.read_csv(r'JPM.csv', index_col=0,parse_dates=True)

# Lamda: 1600
Close_cycle,Close_trend = hpfilter(df['Close'], lamb=1600)
Close_trend.plot(figsize=(15,6)).autoscale(axis='x',tight=True)
Close_cycle.plot(figsize=(15,6)).autoscale(axis='x',tight=True)
```

```python
# Detrending using inbuilt pandas diff()
diff = df['Close'].diff()

plt.figure(figsize=(15,6))
plt.plot(diff)
plt.title('Detrending using Pandas Differencing', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Close Price')
plt.show()
```

```python
# Detrending using Scipy Signal
from scipy import signal

detrended = signal.detrend(df['Close'].values)

plt.figure(figsize=(15,6))
plt.plot(detrended)
plt.xlabel('Close')
plt.ylabel('Frequency')
plt.title('Detrending using Scipy Signal', fontsize=16)
plt.show()
```

```python
# Detrending using HP Filter
from statsmodels.tsa.filters.hp_filter import hpfilter

Close_cycle, Close_trend = hpfilter(df['Close'], lamb=1600)
df['trend'] = Close_trend
detrended = df['Close'] - df['trend']

plt.figure(figsize=(15,6))
plt.plot(detrended)
plt.title('Detrending using HP Filter', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Close Price')
plt.show()
```

## Detrending using Pandas Differencing

* Default Deferencing
* Two-day Deferencing
* Column Deferencing

[Link For Differencing Colab by Mam](https://colab.research.google.com/drive/195zB33lRNrJksmVfydzHBQ8MpL2Q7phZ?usp=sharing)

```python
# Using inbuilt pandas diff()
df = pd.read_csv(r'India_Exchange_Rate_Dataset.csv', index_col=0,parse_dates=True)
diff = df.EXINUS.diff()

plt.figure(figsize=(15,6))
plt.plot(diff)
plt.title('Detrending using Pandas Differencing', fontsize=16)
plt.xlabel('Year')
plt.ylabel('EXINUS exchange rate')
plt.show()
```

```python
# Manually without .diff() function
series = pd.read_csv(r'India_Exchange_Rate_Dataset.csv', index_col=0,parse_dates=True)
X = series.values
diff = list()
for i in range(1, len(X)):
  value = X[i] - X[i - 1]
  diff.append(value)

plt.figure(figsize=(15,6))
plt.plot(diff)
plt.title('Detrending using Manual Differencing', fontsize=16)
plt.xlabel('Year')
plt.ylabel('EXINUS exchange rate')
plt.show()
```

## Detrending using SciPy Signal

```python
from scipy import signal

df = pd.read_csv(r'India_Exchange_Rate_Dataset.csv', index_col=0,parse_dates=True)
detrended = signal.detrend(df.EXINUS.values)

plt.figure(figsize=(15,6))
plt.plot(detrended)
plt.xlabel('EXINUS')
plt.ylabel('Frequency')
plt.title('Detrending using Scipy Signal', fontsize=16)
plt.show()
```

## Pandas Differencing using diff() and from scratch

```python
df = pd.read_csv('POP.csv')
diff = df.value.diff()

plt.figure(figsize=(15,6))
plt.plot(diff)
plt.title('Detrending using Pandas Differencing', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Value')
plt.show()
```

```python
# From Scratch

series = pd.read_csv(r'POP.csv', index_col=0,parse_dates=True)

X = series.value

diff = list()
for i in range(1, len(X)):
  value = int(X[i]) - int(X[i - 1])
  diff.append(value)

plt.figure(figsize=(15,6))
plt.plot(diff)
plt.title('Detrending using Pandas Differencing from Scratch', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Value')
plt.show()
```

## SciPy Signal.detrend method

```python
from scipy import signal

df = pd.read_csv('POP.csv')
detrended = signal.detrend(df.value.values)

plt.figure(figsize=(15,6))
plt.plot(detrended)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Detrending using Scipy Signal', fontsize=16)
plt.show()
```

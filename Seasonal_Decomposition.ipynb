```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv("JPM.csv", index_col=0, parse_dates=True)
df.head()
```

```python
close = df[['Close']]

result = seasonal_decompose(close, model='multiplicative', period=1)
result.plot()
```
